## Drop the unneccesory columns present in dataset

```python
df=df.drop(['Timestamp','Country','state','work_interfere','comments','mental_health_interview','phys_health_interview','wellness_program'],axis=1)
```

## Replace all different Gender name into three specific gender

```python
df['Gender']=df['Gender'].replace(['female','F','Cis Female','Femake','woman','f','Woman','Female (cis)','Female ','femail'],'Female')
df['Gender']=df['Gender'].replace(['Male ','M', 'male', 'm', 'Male-ish', 'maile','Cis Male','Mal','Male (CIS)','male leaning androgynous','Make''Male ', 'Man','cis male','Malr','ostensibly male'],'Male')
df['Gender']=df['Gender'].replace(['Trans-female', 'something kinda male?','queer/she/they', 'non-binary', 'Make', 'Nah', 'All', 'Enby','fluid', 'Genderqueer','Androgyne', 'Agender','cis-female/femme', 'Guy (-ish) ^_^','Trans woman','msle', 'Neuter', 'Female (trans)', 'queer', 'Mail','A little about you', 'p', 'Cis Man','ostensibly male, unsure what that really means' ],'Others')
```

## Perform encoding(convet categorical data into the numerical data) in all column expect age column

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()

UnEncoded_col=df.iloc[:, 1:]
for col in UnEncoded_col:
    df[col]=label_encoder.fit_transform(df[col])
```

## split the independent and dependent data

```python
x=df.drop('treatment',axis=1)
y=df['treatment']
```

## Perform training and testing of dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=29)
```

## Check the score of testing ,training of model and accuracy by using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
training_score=model.score(x_train,y_train)
testing_score=model.score(x_test,y_test)

print("Training score : ",training_score)
print("Testing score : ",testing_score)

predictions=model.predict(x_test)
print(classification_report(y_test,predictions))

print("Accuracy : ",accuracy_score(y_test,predictions))
```
